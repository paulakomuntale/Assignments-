```python
# Simple Demo of Reinforcement Learning
# Import Random
# Give reward to a Robot for reaching the right goal interns of direction
import random
# Enironment
goal = 5
position = 0
# Output
print(' Robot starts at postion: ', position)
while position != goal:
# Action, Random action
action = random.choice(['LEFT', 'RIGHT'])
if action == "RIGHT":
position += 1
else:
position = max(0, position -1)
#Rewards
if position == goal:
reward = 100
else:
reward = -1
# Output the action of the robot
print(f"Action: {action:5} position: {position} Reward: {reward}")
print("\n Robot has Reached our goal" )
```
